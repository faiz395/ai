```python
from queue import PriorityQueue
v = 14

graph = [[] for i in range(v)]

#function for BFS

def bfs(sn,gn, n):
  visited = [0]*n
  pq = PriorityQueue()
  pq.put((0, sn))
  while pq.empty() == False:
    u = pq.get()[1]
    print(u, end = " ")
    if u == gn:
      break
    for m,c in graph[u]:
      if visited[m] == False:
        visited[m] = True
        pq.put((c,m))

#Function for the graph
def addedge(x, y, cost):
  graph[x].append((y,cost))

#creating  graph by callling function
addedge(0,1,3)
addedge(0,2,6)
addedge(0,3,5)
addedge(1,4,9)
addedge(1,5,8)
addedge(2,6,12)
addedge(2,7,14)
addedge(3,8,7)
addedge(8,9,5)
addedge(8,10,6)
addedge(9,11,1)
addedge(9,12,10)
addedge(9,13,2)

bfs(0,13,v)
```
